### EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("./train.csv")
train
```

## 1. categorical variable vs y (credit)
#### 1) gender 
  - 여성과 남성에 따라 신용 점수에 차이가 있는가?
  - 교차분석

```python
from scipy.stats import chi2_contingency
```

- crosstab(교차표) 
  - option : margins=True(all observed), normalize="all"(proportion)

```python
gender_credit = pd.crosstab(train["gender"], train["credit"], margins=True)
print(gender_credit)
```

- chi2_contigency : 교차분석
  - return 카이제곱 통계량, p-value, df(행의 개수-1)*(열의 개수-1), 기댓값

```python
chi2_contingency(observed=gender_credit, correction=False)
```

- 남,여에 따라 신용점수에 차이가 없는 것을 알 수 있다.

#### 2) car - 차량 소유 여부

```python
car_credit = pd.crosstab(train["car"], train["credit"], margins=True)
chi2_contingency(observed=car_credit, correction=False)
```

#### 3) reality - 부동산 소유 여부
- (p-value:0.08) 부동산 소유 여부에 따라 신용 점수에 유의한 차이가 있음을 알 수 있다. 

```python
reality_credit = pd.crosstab(train["reality"], train["credit"], margins=True)
chi2_contingency(observed=reality_credit, correction=False)
```

#### 4) income_type - 소득 분류
- commercial associate : 상업 회원
- pensioner : 연금 수령자
- state servant : 국가 공무원
- student : 학생
- working : 일, 근무

```python
income_type_credit = pd.crosstab(train["income_type"], train["credit"], margins=True)
income_type_credit
```

```python
chi2_contingency(observed=income_type_credit, correction=False)
```
